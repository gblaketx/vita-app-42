```python
#Imports
import json, os, datetime
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from collections import Counter
os.chdir("C:\Users\gblak\OneDrive\CodePractice\webdev\\vita-app\data")
```

```python
with open("parsed_entries.json", "r") as f:
    entries = json.load(f)

with open("health.json", "r") as f:
    walks = json.load(f)
    
totalWordCounts = Counter()
for entry in entries:
    for key, val in entry["wordCounts"].iteritems():
        totalWordCounts[key] += val["count"]
        
mostCommon = set()
for item in totalWordCounts.most_common(500):
    mostCommon.add(item[0])
    
mostCommonList = list(mostCommon)
```

```python
data = np.empty((len(entries), len(mostCommonList) + 1))
walksDict = {}
for walk in walks:
    date = walk["date"].split('T')[0]
    date = datetime.datetime.strptime(date, "%Y-%m-%d")
    walksDict[date] = walk["distance"]
#     print date.day, date.month, date.year
```

```python
n = len(mostCommonList)
for i in range(len(entries)):
    entry = entries[i]
    for j in range(n):
        word = mostCommonList[j]
        if word in entry["wordCounts"]:
            data[i, j] = entry["wordCounts"][word]["count"]
        else:
            data[i, j] = 0
    date = datetime.datetime.strptime(
        entry["timestamp"].split('T')[0], "%Y-%m-%d")
    data[i, n] = walksDict[date]
```

```python
regr = linear_model.LinearRegression()
data_x = data[:len(entries),:n]
data_y = data[:len(entries),n]
regr.fit(data_x, data_y)
coef = regr.coef_
print('Coefficients: \n', regr.coef_)
```

```python
maxIndex = np.argmax(coef)
data_max = data[:len(entries), maxIndex]
print maxIndex, coef[maxIndex], mostCommonList[maxIndex]
```

```python
# Plot outputs
regr_max = linear_model.LinearRegression()
regr_max.fit(data_max.reshape(-1,1), data_y)
data_max_pred = regr_max.predict(data_max.reshape(-1, 1))

plt.scatter(data_max, data_y,  color='black')
plt.plot(data_max, data_max_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
sentiment = np.empty(len(entries))
entryLength = np.empty(len(entries))
distance = np.empty(len(entries))

for i in range(len(entries)):
    entry = entries[i]
    sentiment[i] = entry["sentiment"]["score"]
    entryLength[i] = entry["length"]
    date = datetime.datetime.strptime(
        entry["timestamp"].split('T')[0], "%Y-%m-%d")
    distance[i] = walksDict[date]
```

```python
regr_sent = linear_model.LinearRegression()
regr_sent.fit(distance.reshape(-1, 1), sentiment)
coef = regr_sent.coef_
print('Coefficients: \n', coef)
```

```python

sent_pred = regr_sent.predict(distance.reshape(-1, 1))

plt.scatter(distance, sentiment,  color='black')
plt.plot(distance, sent_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
